# Analisando os dados de entrada

```python
df_train.info()
```

### A classe dos passageiros influencia sua taxa de sobrevivencia?

```python
df_train[['Pclass', 'Survived']].groupby(['Pclass'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

### O sexo dos passageiros influencia sua taxa de sobrevivencia?

```python
df_train[["Sex", "Survived"]].groupby(['Sex'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

### A idade dos passageiros influencia sua taxa de sobrevivencia?

```python
df_train[["Age", "Survived"]].groupby(["Age"], as_index=False).mean().sort_values(by='Survived', ascending=False)
```
